```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

driver = webdriver.Chrome()

url = 'https://www.forbes.com/sites/chasewithorn/2024/04/02/forbes-worlds-billionaires-list-2024-the-top-200/?sh=67b3016430a7'

driver.get(url)

time.sleep(30)

links = driver.find_elements(By.XPATH, '//a[@class="color-link"]')[2:]

billionaire_list = []

for index in range(len(links)):
    links = driver.find_elements(By.XPATH, '//a[@class="color-link"]')[2:]
    l = links[index].get_attribute('href')

    time.sleep(10)
    
    driver.get(l)

    time.sleep(30)

    try: 
        rank = driver.find_element(By.XPATH, '//div[@class="listuser-item__list--rank"]').text.strip()
        name = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/h1'))).text.strip()
        title = ''
        try:
            title_element = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[2]/div')))
            title = title_element.text.strip()
        except Exception as e:
            title = 'N/A'
        try:
            organization = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[2]/div/a'))).text.strip()
        except Exception as e:
            pass
            try:
                organization = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[2]/div/span'))).text.strip()
            except Exception as e:
                organization = 'N/A'
        try:
            net_worth = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[3]/div/div/div[1]/div[1]/div'))).text.strip()
        except Exception as e:
            pass
            try:
                net_worth = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[3]/div/div[1]/div'))).text.strip()
            except Exception as e:
                pass
        bio = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[2]/div[1]/div[1]/ul'))).text.strip()

        stack_data = {}

        title_elements = driver.find_elements(By.CLASS_NAME, 'profile-stats__title')
        text_elements = driver.find_elements(By.XPATH, '//span[@class="profile-stats__text"]')

        for title_element, text_element in zip(title_elements, text_elements):
            titles = title_element.text.strip()
            text = text_element.text.strip()
            stack_data[titles] = text

    except Exception as e:
        print(f"Error{e}")
        
    billionaire_list.append({
        'Rank': rank, 'Name': name, 'Title': title, 'Organization': organization, 'Networth': net_worth, 'Stack': stack_data, 'Bio': bio
    })

    print(billionaire_list[-1])

    df = pd.DataFrame(billionaire_list)
    #df.to_excel(r"C:\Users\ikana\billionaires_list.xlsx", index=False)

    time.sleep(10)

    driver.back()
```

```python
details_expanded = pd.json_normalize(df['Stack'])

forbes_df = pd.concat([df.drop(columns=['Stack']), details_expanded], axis=1)
```

```python
df['Networth'] = df['Networth'].str.replace(r'[$,]', '', regex=True).replace({'B': 'e9', 'M': 'e6'}, regex=True).astype(float)
```

```python
forbes_df["Age"] = forbes_df["Age"].astype("int")

columns_to_convert = ['Children', 'Self-Made Score', 'Philanthropy Score']
for column in columns_to_convert:
    forbes_df[column] = forbes_df[column].astype("float")
```

```python
forbes_df.info()
```

### Self-Made Score vs. Net Worth

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Group by Self-Made Score and compute average Networth
grouped = forbes_df.groupby('Self-Made Score')['Networth'].mean().reset_index()
print("Average Networth by Self-Made Score:")
print(grouped)

# Optional: Plot average Networth by Self-Made Score
plt.figure(figsize=(8, 6))
plt.bar(grouped['Self-Made Score'], grouped['Networth'], color='orange', alpha=0.7)
plt.title('Average Networth by Self-Made Score')
plt.xlabel('Self-Made Score')
plt.ylabel('Average Networth (in billions)')
plt.show()
```

```python
# Box Plot
plt.figure(figsize=(8, 6))
forbes_df.boxplot(column='Networth', by='Self-Made Score', grid=False)
plt.title('Networth Distribution by Self-Made Score')
plt.suptitle('')  # Remove the automatic super title
plt.xlabel('Self-Made Score')
plt.ylabel('Networth (in billions)')
plt.show()
```

### Net Worth vs. Age Analysis

```python
# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(forbes_df['Age'], forbes_df['Networth'], color='green', alpha=0.7)
plt.title('Net Worth vs Age')
plt.xlabel('Age')
plt.ylabel('Net Worth (in billions)')
plt.grid(True)
plt.show()

# Correlation Coefficient
correlation = np.corrcoef(forbes_df['Age'], forbes_df['Networth'])[0, 1]
print(f"Correlation Coefficient between Age and Net Worth: {correlation:.2f}")
```

A correlation coefficient of **-0.27** indicates a weak negative relationship between age and net worth in the dataset. 

This means that as age increases, net worth tends to decrease slightly, but the relationship is not strong or consistent. 

Other factors, such as the source of wealth or self-made score, might play a more significant role in determining net worth.
