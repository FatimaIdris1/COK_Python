```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd

# Define the Forbes URL
forbes_url = "https://www.forbes.com/real-time-billionaires/#12153a773d78"

# Fetch the data from the URL
response = requests.get(forbes_url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    soup = bs(response.text, 'html.parser')

    # Find the table on the page
    table = soup.find('table')
    if table:
        # Extract table headers
        headers = [header.text.strip() for header in table.find_all('th')]

        # Create a DataFrame with the headers
        df = pd.DataFrame(columns=headers)

        # Extract table rows (excluding header row)
        rows = table.find_all('tr')[1:]  # Skip the first row (headers)

        # Iterate over table rows and populate the DataFrame
        for row in rows:
            cells = row.find_all('td')
            row_data = [cell.text.strip() for cell in cells]

            if row_data:  # Ensure the row has data
                df.loc[len(df)] = row_data

        # Output the DataFrame
        print(df)
    else:
        print("No table found on the page.")
else:
    print(f"Failed to fetch the page. Status code: {response.status_code}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

driver = webdriver.Chrome()
  
url = 'https://www.forbes.com/real-time-billionaires/#9b4944e3d788'

driver.get(url)

# Wait for the page to load completely
time.sleep(30)  # Adjust the delay as needed based on page load time

try:
    # Locate the specific table using the CSS selector
    table = driver.find_element(By.CSS_SELECTOR, "div.table-parent.ng-scope.ng-isolate-scope > div > div.fbs-table > div.scrolly-table > table")

    # Extract headers
    headers = [header.text.strip() for header in table.find_elements(By.TAG_NAME, "th")]
    print(f"Headers: {headers}")

    # Extract rows
    rows = []
    for row in table.find_elements(By.TAG_NAME, "tr")[1:]:  # Skip the header row
        cols = [col.text.strip() for col in row.find_elements(By.TAG_NAME, "td")]

        # Pad or trim rows to match the number of headers
        if len(cols) < len(headers):
            cols.extend([None] * (len(headers) - len(cols)))  # Pad with None
        elif len(cols) > len(headers):
            cols = cols[:len(headers)]  # Trim extra columns

        rows.append(cols)

    # Convert to a Pandas DataFrame
    df = pd.DataFrame(rows, columns=headers)
    print(df)

    # Save the table to a CSV file
    df.to_csv("specific_table.csv", index=False)
    print("Table data saved to 'specific_table.csv'")

except Exception as e:
    print(f"Error extracting table: {e}")

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

driver = webdriver.Chrome()

url = 'https://www.forbes.com/sites/chasewithorn/2024/04/02/forbes-worlds-billionaires-list-2024-the-top-200/?sh=67b3016430a7'

driver.get(url)

time.sleep(30)

links = driver.find_elements(By.XPATH, '//a[@class="color-link"]')[2:]

billionaire_list = []

for index in range(len(links)):
    links = driver.find_elements(By.XPATH, '//a[@class="color-link"]')[2:]
    l = links[index].get_attribute('href')

    time.sleep(10)
    
    driver.get(l)

    time.sleep(30)

    try: 
        rank = driver.find_element(By.XPATH, '//div[@class="listuser-item__list--rank"]').text.strip()
        name = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/h1'))).text.strip()
        title = ''
        try:
            title_element = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[2]/div')))
            title = title_element.text.strip()
        except Exception as e:
            title = 'N/A'
        try:
            organization = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[2]/div/a'))).text.strip()
        except Exception as e:
            pass
            try:
                organization = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[2]/div/span'))).text.strip()
            except Exception as e:
                organization = 'N/A'
        try:
            net_worth = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[3]/div/div/div[1]/div[1]/div'))).text.strip()
        except Exception as e:
            pass
            try:
                net_worth = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[1]/div[2]/div/div/div/div[1]/div[3]/div/div[1]/div'))).text.strip()
            except Exception as e:
                pass
        bio = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '/html/body/div[1]/div[1]/div[2]/div[1]/div[1]/ul'))).text.strip()

        stack_data = {}

        title_elements = driver.find_elements(By.CLASS_NAME, 'profile-stats__title')
        text_elements = driver.find_elements(By.XPATH, '//span[@class="profile-stats__text"]')

        for title_element, text_element in zip(title_elements, text_elements):
            titles = title_element.text.strip()
            text = text_element.text.strip()
            stack_data[titles] = text

    except Exception as e:
        print(f"Error{e}")
        
    billionaire_list.append({
        'Rank': rank, 'Name': name, 'Title': title, 'Organization': organization, 'Networth': net_worth, 'Stack': stack_data, 'Bio': bio
    })

    print(billionaire_list[-1])

    df = pd.DataFrame(billionaire_list)
    #df.to_excel(r"C:\Users\ikana\billionaires_list.xlsx", index=False)

    time.sleep(10)

    driver.back()
```
